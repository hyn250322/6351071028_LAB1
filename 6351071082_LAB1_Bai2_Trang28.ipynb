```python
# Câu 1: Nhập dữ liệu và thống kê thông tin cơ bản
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
from matplotlib import pyplot as plt
from matplotlib import style
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)


# Đọc dữ liệu (giả sử file có tên bank_marketing.csv)
df = pd.read_csv(
    'bank.csv',
    dtype={'duration': 'int64', 'cons.price.idx': 'float64'}
)

print("Họ tên: Huỳnh Thiên Huy - MSSV: 6351071028")

# Hiển thị 12 dòng đầu
df.head(12)

```

```python
# Câu 2: Liệt kê 5 thuộc tính có số lượng giá trị rỗng nhiều nhất
total = df.isnull().sum().sort_values(ascending=False)
percent_1 = df.isnull().sum() / df.isnull().count() * 100
percent_2 = (round(percent_1, 1)).sort_values(ascending=False)
missing_data = pd.concat([total, percent_2], axis=1, keys=['Total', '%'])
print("Họ tên: Huỳnh Thiên Huy - MSSV: 6351071028")
missing_data.head(5)
```

```python
# Câu 3: Vẽ biểu đồ khảo sát mối liên quan giữa Age, Job, Education và Y (Subscribe)
print("Họ tên: Huỳnh Thiên Huy - MSSV: 6351071028")

# Tạo biến nhị phân cho y
df['y_binary'] = df['y'].map({'yes': 1, 'no': 0})

# Biểu đồ 1: Phân bố tuổi theo kết quả đăng ký
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))

# 1. Phân bố tuổi theo Y
subscribed = 'subscribed (yes)'
not_subscribed = 'not subscribed (no)'

ax = sns.histplot(
    df[df['y'] == 'yes']['age'].dropna(),
    bins=20,
    label=subscribed,
    ax=axes[0, 0],
    kde=True,
    color='g',
    alpha=0.6
)
sns.histplot(
    df[df['y'] == 'no']['age'].dropna(),
    bins=20,
    label=not_subscribed,
    ax=axes[0, 0],
    kde=True,
    color='r',
    alpha=0.6
)
ax.legend()
ax.set_title('Phân bố độ tuổi theo kết quả đăng ký', fontsize=12, fontweight='bold')
ax.set_xlabel('Tuổi')
ax.set_ylabel('Số lượng')

# 2. Tỷ lệ đăng ký theo Job
job_subscribe = df.groupby('job')['y_binary'].agg(['sum', 'count'])
job_subscribe['rate'] = (job_subscribe['sum'] / job_subscribe['count']) * 100
job_subscribe = job_subscribe.sort_values('rate', ascending=False)

sns.barplot(
    x=job_subscribe.index,
    y=job_subscribe['rate'],
    ax=axes[0, 1],
    palette='viridis'
)
axes[0, 1].set_title('Tỷ lệ đăng ký theo Nghề nghiệp', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Nghề nghiệp')
axes[0, 1].set_ylabel('Tỷ lệ đăng ký (%)')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Tỷ lệ đăng ký theo Education
edu_subscribe = df.groupby('education')['y_binary'].agg(['sum', 'count'])
edu_subscribe['rate'] = (edu_subscribe['sum'] / edu_subscribe['count']) * 100
edu_subscribe = edu_subscribe.sort_values('rate', ascending=False)

sns.barplot(
    x=edu_subscribe.index,
    y=edu_subscribe['rate'],
    ax=axes[1, 0],
    palette='mako'
)
axes[1, 0].set_title('Tỷ lệ đăng ký theo Trình độ học vấn', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Trình độ học vấn')
axes[1, 0].set_ylabel('Tỷ lệ đăng ký (%)')
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Boxplot Age theo Y
sns.boxplot(
    data=df,
    x='y',
    y='age',
    ax=axes[1, 1],
    palette='Set2'
)
axes[1, 1].set_title('Phân bố tuổi theo kết quả đăng ký', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Kết quả đăng ký')
axes[1, 1].set_ylabel('Tuổi')

plt.tight_layout()
plt.show()
```

```python
# Câu 4: Vẽ biểu đồ khảo sát mối liên quan giữa Loan, Housing, Marital và Y
print("Họ tên: Huỳnh Thiên Huy - MSSV: 6351071028")

# FacetGrid cho Loan và Marital status
FacetGrid = sns.FacetGrid(
    df,
    row='loan',
    col='marital',
    height=4,
    aspect=1.2,
    margin_titles=True
)

FacetGrid.map(
    sns.countplot,
    'y',
    order=['yes', 'no'],
    palette='Set2'
)

FacetGrid.add_legend()
FacetGrid.fig.subplots_adjust(top=0.92)
FacetGrid.fig.suptitle('Mối quan hệ giữa Loan, Marital Status và Y (Subscribe)', fontsize=14, fontweight='bold')
plt.show()

# Biểu đồ chi tiết
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 12))

# 1. Housing và Y
sns.countplot(
    data=df,
    x='housing',
    hue='y',
    ax=axes[0, 0],
    palette='pastel'
)
axes[0, 0].set_title('Mối quan hệ giữa Housing Loan và Subscribe', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Housing Loan')
axes[0, 0].set_ylabel('Số lượng')
axes[0, 0].legend(title='Subscribe', labels=['No', 'Yes'])

# 2. Personal Loan và Y
sns.countplot(
    data=df,
    x='loan',
    hue='y',
    ax=axes[0, 1],
    palette='pastel'
)
axes[0, 1].set_title('Mối quan hệ giữa Personal Loan và Subscribe', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Personal Loan')
axes[0, 1].set_ylabel('Số lượng')
axes[0, 1].legend(title='Subscribe', labels=['No', 'Yes'])

# 3. Marital Status và Y
marital_y = pd.crosstab(df['marital'], df['y'], normalize='index') * 100
marital_y.plot(kind='bar', ax=axes[1, 0], color=['coral', 'lightgreen'])
axes[1, 0].set_title('Tỷ lệ đăng ký theo Tình trạng hôn nhân', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Tình trạng hôn nhân')
axes[1, 0].set_ylabel('Tỷ lệ (%)')
axes[1, 0].legend(title='Subscribe', labels=['No', 'Yes'])
axes[1, 0].tick_params(axis='x', rotation=45)

# 4. Default và Y
sns.countplot(
    data=df,
    x='default',
    hue='y',
    ax=axes[1, 1],
    palette='Set3'
)
axes[1, 1].set_title('Mối quan hệ giữa Credit Default và Subscribe', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Credit Default')
axes[1, 1].set_ylabel('Số lượng')
axes[1, 1].legend(title='Subscribe', labels=['No', 'Yes'])

plt.tight_layout()
plt.show()

```

```python
# Câu 5: Thống kê đơn giản về Duration và Campaign
print("Họ tên: Huỳnh Thiên Huy - MSSV: 6351071028")

# Tạo biến nhị phân
df['y_binary'] = df['y'].map({'yes': 1, 'no': 0})

# Vẽ 2 biểu đồ cơ bản
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# 1. Duration và Y
sns.boxplot(data=df, x='y', y='duration', ax=axes[0], palette='Set3')
axes[0].set_title('Duration và Subscribe', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Subscribe')
axes[0].set_ylabel('Duration (giây)')

# 2. Campaign và Y
sns.boxplot(data=df, x='y', y='campaign', ax=axes[1], palette='Set2')
axes[1].set_title('Campaign và Subscribe', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Subscribe')
axes[1].set_ylabel('Số lần contact')

plt.tight_layout()
plt.show()

# Thống kê cơ bản
print("\nThống kê Duration:")
print(df.groupby('y')['duration'].describe()[['mean', 'min', 'max']])

print("\nThống kê Campaign:")
print(df.groupby('y')['campaign'].describe()[['mean', 'min', 'max']])

```

```python
# Câu 6: Tiền xử lý dữ liệu thiếu và xóa bỏ thuộc tính không cần thiết
print("Họ tên: Huỳnh Thiên Huy - MSSV: 6351071028")

print("="*70)
print("TIỀN XỬ LÝ DỮ LIỆU")
print("="*70)

# 1. Kiểm tra giá trị 'unknown'
print("\n1. Kiểm tra giá trị 'unknown' trong các cột:")
unknown_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'poutcome']
for col in unknown_cols:
    if col in df.columns:
        unknown_count = (df[col] == 'unknown').sum()
        unknown_pct = (unknown_count / len(df)) * 100
        print(f"   - {col}: {unknown_count} ({unknown_pct:.2f}%)")

# 2. Xử lý 'unknown' values
print("\n2. Xử lý giá trị 'unknown':")
for col in unknown_cols:
    if col in df.columns:
        unknown_count = (df[col] == 'unknown').sum()

        if unknown_count > 0:
            # Với cột có ít 'unknown', thay bằng mode
            if unknown_count < len(df) * 0.1:  # Dưới 10%
                mode_value = df[df[col] != 'unknown'][col].mode()
                if len(mode_value) > 0:
                    mode_value = mode_value[0]
                    df[col] = df[col].replace('unknown', mode_value)
                    print(f"   {col}: Đã thay 'unknown' bằng mode '{mode_value}'")
            else:
                # Với cột có nhiều 'unknown', giữ nguyên hoặc tạo category mới
                print(f"   ! {col}: Giữ nguyên 'unknown' (tỷ lệ cao)")

# 3. Kiểm tra giá trị null
print("\n3. Kiểm tra giá trị NULL:")
null_counts = df.isnull().sum()
if null_counts.sum() > 0:
    print(null_counts[null_counts > 0])
else:
    print("   Không có giá trị NULL")

# 4. Xử lý các giá trị null (nếu có)
numeric_cols = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']
for col in numeric_cols:
    if col in df.columns and df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"   {col}: Đã fill NULL bằng median = {median_val}")

# 5. Xử lý cột pdays (giá trị -1 nghĩa là chưa từng contact)
if 'pdays' in df.columns:
    print("\n4. Xử lý cột 'pdays':")
    print(f"   - Số lượng pdays = -1 (chưa từng contact): {(df['pdays'] == -1).sum()}")
    # Có thể tạo biến mới để đánh dấu khách hàng mới
    df['was_contacted_before'] = (df['pdays'] != -1).astype(int)
    print("   Đã tạo biến 'was_contacted_before'")

# 6. Xóa các cột tạm thời
print("\n5. Xóa các cột tạm thời:")
temp_cols = ['duration_group', 'campaign_group', 'previous_group', 'y_binary']
for col in temp_cols:
    if col in df.columns:
        df = df.drop([col], axis=1)
        print(f"   Đã xóa cột '{col}'")

# 7. Kiểm tra outliers cho các cột số
print("\n6. Kiểm tra Outliers:")
for col in ['age', 'balance', 'duration', 'campaign']:
    if col in df.columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outliers_pct = (outliers / len(df)) * 100
        print(f"   - {col}: {outliers} outliers ({outliers_pct:.2f}%)")

# 8. Tóm tắt sau xử lý
print("\n" + "="*70)
print("TÓM TẮT SAU XỬ LÝ:")
print("="*70)
print(f"Số lượng mẫu: {len(df)}")
print(f"Số lượng thuộc tính: {len(df.columns)}")
print(f"\nCác thuộc tính còn lại:")
print(df.columns.tolist())

print("\nThông tin chi tiết:")
df.info()

print("\nMẫu dữ liệu sau xử lý:")
print(df.head(10))

# 9. Lưu dữ liệu đã xử lý (optional)
# df.to_csv('bank_marketing_cleaned.csv', index=False)
# print("\n Đã lưu dữ liệu đã xử lý vào 'bank_marketing_cleaned.csv'")
```
